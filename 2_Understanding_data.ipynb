##  B) Heart Disease Dataset

Below is the scatter plot that describes distribution and correlation between the features of the dataset namely age, resting blood pressure(trestbps), cholesterol(chol), chest pain(cp) and target(presence or absence of disease).

Here are my observation from the plot:

- i) From histogram, people in early 40's and early 50's tend to have more heart disease as compared to people in mid 50' to late 60's
- ii) People who have experienced either atypical angina and non-angina (types of chest pain) are more likely to suffer from heart disease than people who have experienced asymptomatic angina and typical aniga.
- iii) Age and resting blood-pressure have positive correlation indicating increase in age results in increase in blood pressure. Also the people having resting blood pressure greater than 130 are suffering from heart disease.
- iv) People with the cholesterol less than 170 don't suffer from heart disease however with cholesterol level ranging between 200 to 250 are suffering from heart diseases. 

- v) There is not much correlation among the features of the dataset meaning that there is no any redundancy among the features and hence each feature depicts higher level of independence for the target variable.

- i) Age : When it comes to heart disease age tends to play a pivotal rode. Greater the age more is the risk of suffering from heart ailment. 
- ii) Resting blood pressure: There is always a strong link between blood pressure and heart disease. When the BP goes high it might rupture the arteries carrying blood resulting in heart stroke.
- iii) Cholesterol : Increasing cholesterol level leads to blocking of arteries carrying blood and can further lead to increase problems relating to heart.
- iv) Chest pain : Most the time people suffer from chest pain due to heart muscles not getting enough oxygen and which may further lead to heart stroke.<br>

All this features independently play pivotal in deciding whether person suffers from heart disease or not.
