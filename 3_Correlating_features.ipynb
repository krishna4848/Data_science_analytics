```python
# AUTHOR : KEVAL PRAJAPATI
# AIM : DATA CORRELATION

# Correlation Co-efficent for all the features of Iris Dataset
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

iris_df = pd.read_csv('D:\COLLEGE\Waterloo\TERM 1\ECE 657A\Assignment 1\dataset\iris_dataset_missing_filled_final.csv')


# Calulating correlation co-efficient for the all the features namely sepal length, sepal width, petal length, petal width
print('\n Calulating correlation co-efficient for the all the features namely sepal length, sepal width, petal length, petal width:\n')
cor_fea= iris_df.corr()
print(cor_fea)

# Using heat map for easier understanding
print("\n Heat map of correlation matrix:\n")
m = np.triu(cor_fea)
sns.heatmap(cor_fea, annot = True, fmt='.3g',cmap = 'seismic',square = True)


# Statistics about the features Mean, variance, skew, kurtosis

#1) Mean: 
print("\nMean:\n")
print(iris_df.mean())

#2) Variance: 
print("\nVariance:\n")
print(iris_df.var())

#3) Skew: 
print("\nSkew:\n")
print(iris_df.skew())

#4) Kurtosis: 
print("\nKurtosis:\n")
print(iris_df.kurtosis())
```
